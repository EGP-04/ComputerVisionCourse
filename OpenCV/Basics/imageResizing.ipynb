```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/emmanuelgeorgep/Documents/Courses/ComputerVision/OpenCV/images/giraffe-Kenya.png")
img.shape
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
new_height,new_width = 400,400

resize_img = cv2.resize(img_rgb,(new_height,new_width))
plt.imshow(resize_img)
```

```python
scale = 0.5
s_width = int(img_rgb.shape[0]*scale)
s_height = int(img_rgb.shape[1]*scale)

scale_img = cv2.resize(img_rgb,(s_width,s_height),interpolation=cv2.INTER_AREA)

print(scale_img.shape)

plt.imshow(scale_img)
```

```python
comp_img = cv2.resize(img_rgb,(100,100))

resize_width = 600
resize_height = 600

methods = {
    "Nearest": cv2.INTER_NEAREST,
    "Linear" : cv2.INTER_LINEAR,
    "Cubic" : cv2.INTER_CUBIC,
    "Area"  : cv2.INTER_AREA,
    "Lanczos4":cv2.INTER_LANCZOS4
}

resized_imgs = {}

for name,method in methods.items():
    resized_imgs[name] = cv2.resize(comp_img,(resize_width,resize_height),interpolation = method)
    
plt.figure(figsize = (12,8))

plt.subplot(2, 3, 1)
plt.imshow(comp_img)
plt.title("Original")
plt.axis('off')

for i, (name, resized_image) in enumerate(resized_imgs.items(), start=2):
    plt.subplot(2,3,i)
    plt.imshow( resized_image)
    plt.title(name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
