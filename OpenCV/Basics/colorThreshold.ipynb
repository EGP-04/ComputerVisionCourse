```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/emmanuelgeorgep/Documents/Courses/ComputerVision/OpenCV/images/giraffe-Kenya_1_optimized.png")
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_rgb)
```

```python
plt.imshow(img_hsv)
```

```python
lower_white = np.array([0,0,190])
upper_white = np.array([30,80,255])
```

```python
mask = cv2.inRange(img_hsv,lower_white,upper_white)
plt.imshow(mask)
```

```python
black_hsv = np.uint8([[[120,255,0]]])
black_bgr = cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)
```

```python
img_res = img.copy()
img_res[mask>0] = black_bgr
```

```python
img_res_rgb = cv2.cvtColor(img_res,cv2.COLOR_BGR2RGB)
```

```python
fig,axis = plt.subplots(1,3,figsize=(12,6))

axis[0].imshow(img_rgb)
axis[0].set_title("Original image")
axis[0].axis('off')

axis[1].imshow(mask)
axis[1].set_title("Mask")
axis[1].axis('off')

axis[2].imshow(img_res)
axis[2].set_title("Resultant Image")
axis[2].axis('off')

plt.tight_layout()
plt.show()
```
