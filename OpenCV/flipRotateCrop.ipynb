```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/emmanuelgeorgep/Documents/Courses/ComputerVision/OpenCV/images/giraffe-Kenya.png")
img.shape
```

```python
#1 - horizontal
#0 - vertical
#-1 - both axes

flip_hor = cv2.flip(img,1)
flip_ver = cv2.flip(img,0)
flip_both = cv2.flip(img,-1)

imgs = {
    "Original" : cv2.cvtColor(img,cv2.COLOR_BGR2RGB),
    "Horizontal" : cv2.cvtColor(flip_hor,cv2.COLOR_BGR2RGB),
    "Vertical" : cv2.cvtColor(flip_ver,cv2.COLOR_BGR2RGB),
    "Both Axes" : cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)
}

plt.figure(figsize=(12,8))
i = 0
for name,img in imgs.items():
    plt.subplot(2,2,i+1)
    plt.imshow(img)
    plt.title(name)
    plt.axis("Off")
    i += 1

plt.tight_layout()
plt.show()
```

```python
img_rgb = imgs["Original"]

rotate_90_clockwise = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(img_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(img_rgb, cv2.ROTATE_180)


titles = ["Original image", "90 degree clockwise", "90 degree counterclockwise", "180 degree"]
images = [img_rgb, 
          rotate_90_clockwise, 
          rotate_90_counterclockwise, 
          rotate_180
        ]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("Off")
plt.tight_layout()
plt.show()
```

```python
x_start, y_start = 400,300
x_end, y_end = 800, 900

cropped_image = img_rgb[y_start:y_end, x_start:x_end]

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original image")
plt.axis("Off")

plt.subplot(1,2,2)
plt.imshow(cropped_image)
plt.title("Cropped image")
plt.axis("Off")

plt.tight_layout()
plt.show()
```
