```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("/Users/emmanuelgeorgep/Documents/Courses/ComputerVision/OpenCV/images/mountain.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
channels = cv2.split(image_rgb)
colors = ['red', 'green', 'blue']

for channel, color in zip(channels, colors):
    hist = cv2.calcHist([channel], [0], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequence")
plt.show()
```

```python
image_hsv = cv2.cvtColor(image_rgb,cv2.COLOR_RGB2HSV)

channels = cv2.split(image_hsv)
names = ["Hue","Saturation","Value"]

hist = []
for channel,name in zip(channels,names):
    hist.append(cv2.calcHist([channel],[0],None,[255],[0,255]))
    
```
