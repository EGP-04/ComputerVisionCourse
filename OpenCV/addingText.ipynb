```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
height, width = 500, 700
sky_blue = (255, 127, 0)
canvas = np.full((height, width, 3), sky_blue, dtype=np.uint8)
canvas = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

plt.imshow(canvas)
```

```python
text1 = "Hello, OpenCV"
text2 = "I am Emmanuel"
font = cv2.FONT_HERSHEY_COMPLEX

org1 = (200, 100)
org2 = (200,200)
font_scale = 1.5
color = (255, 255, 255)
thickness = 2

cv2.putText(canvas, text1, org1, font, font_scale, color, thickness, cv2.LINE_AA)
cv2.putText(canvas,text2,org2,font,font_scale,color,thickness,cv2.LINE_AA)

plt.imshow(canvas)
```
